## **Aula 5.2: Tipos de Text Splitting - Diferentes Maneiras de Cortar**

---

### **T√°, mas quais s√£o os tipos de splitting?**

Existem v√°rias maneiras de "cortar" um texto, como diferentes tipos de facas na cozinha! üç¥

**üñºÔ∏è Sugest√£o de imagem**: Diferentes tipos de facas cortando diferentes tipos de alimentos

### **1. Character Text Splitter - A Faca Simples**

√â como usar uma **faca de p√£o** - corta em peda√ßos de tamanho fixo, sem se importar muito com o conte√∫do.

```python
# üî™ EXEMPLO 1: CHARACTER TEXT SPLITTER - A FACA SIMPLES

# Criando um splitter que corta por caracteres
character_splitter = CharacterTextSplitter(
    separator="\n",  # Corta nas quebras de linha
    chunk_size=200,   # Tamanho m√°ximo de cada peda√ßo
    chunk_overlap=20, # Sobreposi√ß√£o entre peda√ßos
    length_function=len
)

# Cortando nosso texto
chunks_character = character_splitter.split_text(texto_longo)

print(f"üî™ Character Splitter criou {len(chunks_character)} peda√ßos")
print(f"üìè Tamanho m√©dio: {sum(len(chunk) for chunk in chunks_character) / len(chunks_character):.0f} caracteres")

# Mostrando alguns peda√ßos
print("\nüç∞ Primeiros 3 peda√ßos:")
for i, chunk in enumerate(chunks_character[:3], 1):
    print(f"\n--- Peda√ßo {i} ---")
    print(chunk[:150] + "..." if len(chunk) > 150 else chunk)
```

### **2. Recursive Character Text Splitter - A Faca Inteligente**

√â como usar uma **faca de chef** - corta de forma inteligente, respeitando a estrutura do texto (par√°grafos, frases, etc.).

```python
# üß† EXEMPLO 2: RECURSIVE CHARACTER TEXT SPLITTER - A FACA INTELIGENTE

# Criando um splitter recursivo (mais inteligente)
recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,      # Tamanho m√°ximo
    chunk_overlap=50,    # Sobreposi√ß√£o
    length_function=len,
    separators=["\n\n", "\n", ". ", "! ", "? ", " ", ""]  # Ordem de prioridade
)

# Cortando nosso texto de forma inteligente
chunks_recursive = recursive_splitter.split_text(texto_longo)

print(f"üß† Recursive Splitter criou {len(chunks_recursive)} peda√ßos")
print(f"üìè Tamanho m√©dio: {sum(len(chunk) for chunk in chunks_recursive) / len(chunks_recursive):.0f} caracteres")

# Mostrando alguns peda√ßos inteligentes
print("\nüß† Primeiros 3 peda√ßos (mais inteligentes):")
for i, chunk in enumerate(chunks_recursive[:3], 1):
    print(f"\n--- Peda√ßo {i} ---")
    print(chunk[:200] + "..." if len(chunk) > 200 else chunk)
```

### **3. Token Text Splitter - A Faca de Precis√£o**

√â como usar uma **faca de precis√£o** - corta baseado em tokens (unidades de significado), n√£o apenas caracteres.

```python
# üéØ EXEMPLO 3: TOKEN TEXT SPLITTER - A FACA DE PRECIS√ÉO

# Criando um splitter baseado em tokens
token_splitter = TokenTextSplitter(
    chunk_size=100,      # N√∫mero m√°ximo de tokens
    chunk_overlap=20,    # Sobreposi√ß√£o de tokens
    encoding_name="cl100k_base"  # Encoding do GPT
)

# Cortando nosso texto por tokens
chunks_tokens = token_splitter.split_text(texto_longo)

print(f"üéØ Token Splitter criou {len(chunks_tokens)} peda√ßos")
print(f"üìè Tamanho m√©dio: {sum(len(chunk) for chunk in chunks_tokens) / len(chunks_tokens):.0f} caracteres")

# Mostrando alguns peda√ßos baseados em tokens
print("\nüéØ Primeiros 3 peda√ßos (baseados em tokens):")
for i, chunk in enumerate(chunks_tokens[:3], 1):
    print(f"\n--- Peda√ßo {i} ---")
    print(chunk[:180] + "..." if len(chunk) > 180 else chunk)
```

## **Aula 5.4: Splitting Especializado - Para Diferentes Tipos de Texto**

---

### **T√°, mas e se o texto for Markdown ou HTML?**

Assim como temos facas especiais para diferentes alimentos, temos splitters especiais para diferentes tipos de texto! üç¥

**üñºÔ∏è Sugest√£o de imagem**: Diferentes ferramentas para diferentes tipos de material

```python
# üìù EXEMPLO 4: SPLITTING ESPECIALIZADO

# Texto em Markdown
texto_markdown = """
# T√≠tulo Principal

## Subt√≠tulo 1
Este √© um par√°grafo normal.

### Subt√≠tulo 1.1
Outro par√°grafo aqui.

## Subt√≠tulo 2
Mais conte√∫do aqui.

- Item 1
- Item 2
- Item 3

**Texto em negrito** e *texto em it√°lico*.
"""

# Texto em HTML
texto_html = """
<html>
<head><title>P√°gina de Teste</title></head>
<body>
    <h1>T√≠tulo Principal</h1>
    <p>Este √© um par√°grafo.</p>
    <h2>Subt√≠tulo</h2>
    <p>Outro par√°grafo.</p>
    <ul>
        <li>Item 1</li>
        <li>Item 2</li>
    </ul>
</body>
</html>
"""

print("üìù TEXTOS ESPECIALIZADOS CRIADOS!")
print(f"üìÑ Markdown: {len(texto_markdown)} caracteres")
print(f"üåê HTML: {len(texto_html)} caracteres")
```

```python
# üîß TESTANDO SPLITTERS ESPECIALIZADOS

# Markdown Splitter
markdown_splitter = MarkdownTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

# HTML Splitter
html_splitter = HTMLTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

# Testando Markdown
chunks_md = markdown_splitter.split_text(texto_markdown)
print("üìù MARKDOWN SPLITTER:")
print(f"   üìÑ Peda√ßos criados: {len(chunks_md)}")
for i, chunk in enumerate(chunks_md, 1):
    print(f"   {i}. {chunk.strip()[:50]}...")

print("\n" + "-" * 40)

# Testando HTML
chunks_html = html_splitter.split_text(texto_html)
print("üåê HTML SPLITTER:")
print(f"   üìÑ Peda√ßos criados: {len(chunks_html)}")
for i, chunk in enumerate(chunks_html, 1):
    print(f"   {i}. {chunk.strip()[:50]}...")

print("\nüí° OBSERVA√á√ÉO: Splitters especializados respeitam a estrutura do formato!")
```

### **Desafio do M√≥dulo - Criando um Sistema Completo**

Vamos criar um sistema que escolhe automaticamente o melhor splitter para cada tipo de texto:

```python
# üéØ DESAFIO: SISTEMA INTELIGENTE DE SPLITTING

def escolher_splitter_inteligente(texto):
    """Escolhe o melhor splitter baseado no tipo de texto"""
    
    # Detectando tipo de texto
    if texto.strip().startswith('<html') or '<body>' in texto:
        print("üåê Detectado HTML - usando HTML Splitter")
        return HTMLTextSplitter(chunk_size=200, chunk_overlap=30)
    
    elif '#' in texto[:100] and '\n' in texto:
        print("üìù Detectado Markdown - usando Markdown Splitter")
        return MarkdownTextSplitter(chunk_size=200, chunk_overlap=30)
    
    else:
        print("üìÑ Texto normal - usando Recursive Splitter")
        return RecursiveCharacterTextSplitter(
            chunk_size=200,
            chunk_overlap=30,
            separators=["\n\n", "\n", ". ", "! ", "? ", " ", ""]
        )

# Testando o sistema inteligente
textos_teste = [
    ("Texto normal", "Este √© um texto normal sem formata√ß√£o especial."),
    ("Markdown", "# T√≠tulo\n\nEste √© um texto em markdown."),
    ("HTML", "<html><body><h1>T√≠tulo</h1><p>Texto HTML</p></body></html>")
]

print("üéØ TESTANDO SISTEMA INTELIGENTE DE SPLITTING")
print("=" * 50)

for nome, texto in textos_teste:
    print(f"\nüîç Testando: {nome}")
    splitter = escolher_splitter_inteligente(texto)
    chunks = splitter.split_text(texto)
    print(f"   üìÑ Peda√ßos criados: {len(chunks)}")
    print(f"   üìè Tamanho m√©dio: {sum(len(chunk) for chunk in chunks) / len(chunks):.0f} chars")

print("\nüéâ Sistema inteligente funcionando perfeitamente!")
```
