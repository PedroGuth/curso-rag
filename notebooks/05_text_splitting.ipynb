## **Aula 5.2: Tipos de Text Splitting - Diferentes Maneiras de Cortar**

---

### **TÃ¡, mas quais sÃ£o os tipos de splitting?**

Existem vÃ¡rias maneiras de "cortar" um texto, como diferentes tipos de facas na cozinha! ğŸ´

**ğŸ–¼ï¸ SugestÃ£o de imagem**: Diferentes tipos de facas cortando diferentes tipos de alimentos

### **1. Character Text Splitter - A Faca Simples**

Ã‰ como usar uma **faca de pÃ£o** - corta em pedaÃ§os de tamanho fixo, sem se importar muito com o conteÃºdo.

```python
# ğŸ”ª EXEMPLO 1: CHARACTER TEXT SPLITTER - A FACA SIMPLES

# Criando um splitter que corta por caracteres
character_splitter = CharacterTextSplitter(
    separator="\n",  # Corta nas quebras de linha
    chunk_size=200,   # Tamanho mÃ¡ximo de cada pedaÃ§o
    chunk_overlap=20, # SobreposiÃ§Ã£o entre pedaÃ§os
    length_function=len
)

# Cortando nosso texto
chunks_character = character_splitter.split_text(texto_longo)

print(f"ğŸ”ª Character Splitter criou {len(chunks_character)} pedaÃ§os")
print(f"ğŸ“ Tamanho mÃ©dio: {sum(len(chunk) for chunk in chunks_character) / len(chunks_character):.0f} caracteres")

# Mostrando alguns pedaÃ§os
print("\nğŸ° Primeiros 3 pedaÃ§os:")
for i, chunk in enumerate(chunks_character[:3], 1):
    print(f"\n--- PedaÃ§o {i} ---")
    print(chunk[:150] + "..." if len(chunk) > 150 else chunk)
```

### **2. Recursive Character Text Splitter - A Faca Inteligente**

Ã‰ como usar uma **faca de chef** - corta de forma inteligente, respeitando a estrutura do texto (parÃ¡grafos, frases, etc.).

```python
# ğŸ§  EXEMPLO 2: RECURSIVE CHARACTER TEXT SPLITTER - A FACA INTELIGENTE

# Criando um splitter recursivo (mais inteligente)
recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,      # Tamanho mÃ¡ximo
    chunk_overlap=50,    # SobreposiÃ§Ã£o
    length_function=len,
    separators=["\n\n", "\n", ". ", "! ", "? ", " ", ""]  # Ordem de prioridade
)

# Cortando nosso texto de forma inteligente
chunks_recursive = recursive_splitter.split_text(texto_longo)

print(f"ğŸ§  Recursive Splitter criou {len(chunks_recursive)} pedaÃ§os")
print(f"ğŸ“ Tamanho mÃ©dio: {sum(len(chunk) for chunk in chunks_recursive) / len(chunks_recursive):.0f} caracteres")

# Mostrando alguns pedaÃ§os inteligentes
print("\nğŸ§  Primeiros 3 pedaÃ§os (mais inteligentes):")
for i, chunk in enumerate(chunks_recursive[:3], 1):
    print(f"\n--- PedaÃ§o {i} ---")
    print(chunk[:200] + "..." if len(chunk) > 200 else chunk)
```

### **3. Token Text Splitter - A Faca de PrecisÃ£o**

Ã‰ como usar uma **faca de precisÃ£o** - corta baseado em tokens (unidades de significado), nÃ£o apenas caracteres.

```python
# ğŸ¯ EXEMPLO 3: TOKEN TEXT SPLITTER - A FACA DE PRECISÃƒO

# Criando um splitter baseado em tokens
token_splitter = TokenTextSplitter(
    chunk_size=100,      # NÃºmero mÃ¡ximo de tokens
    chunk_overlap=20,    # SobreposiÃ§Ã£o de tokens
    encoding_name="cl100k_base"  # Encoding do GPT
)

# Cortando nosso texto por tokens
chunks_tokens = token_splitter.split_text(texto_longo)

print(f"ğŸ¯ Token Splitter criou {len(chunks_tokens)} pedaÃ§os")
print(f"ğŸ“ Tamanho mÃ©dio: {sum(len(chunk) for chunk in chunks_tokens) / len(chunks_tokens):.0f} caracteres")

# Mostrando alguns pedaÃ§os baseados em tokens
print("\nğŸ¯ Primeiros 3 pedaÃ§os (baseados em tokens):")
for i, chunk in enumerate(chunks_tokens[:3], 1):
    print(f"\n--- PedaÃ§o {i} ---")
    print(chunk[:180] + "..." if len(chunk) > 180 else chunk)
```

**ğŸ–¼ï¸ SugestÃ£o de imagem**: Diferentes ferramentas para diferentes tipos de material

```python
# ğŸ“ EXEMPLO 4: SPLITTING ESPECIALIZADO

# Texto em Markdown
texto_markdown = """
# TÃ­tulo Principal

## SubtÃ­tulo 1
Este Ã© um parÃ¡grafo normal.

### SubtÃ­tulo 1.1
Outro parÃ¡grafo aqui.

## SubtÃ­tulo 2
Mais conteÃºdo aqui.

- Item 1
- Item 2
- Item 3

**Texto em negrito** e *texto em itÃ¡lico*.
"""

# Texto em HTML
texto_html = """
<html>
<head><title>PÃ¡gina de Teste</title></head>
<body>
    <h1>TÃ­tulo Principal</h1>
    <p>Este Ã© um parÃ¡grafo.</p>
    <h2>SubtÃ­tulo</h2>
    <p>Outro parÃ¡grafo.</p>
    <ul>
        <li>Item 1</li>
        <li>Item 2</li>
    </ul>
</body>
</html>
"""

print("ğŸ“ TEXTOS ESPECIALIZADOS CRIADOS!")
print(f"ğŸ“„ Markdown: {len(texto_markdown)} caracteres")
print(f"ğŸŒ HTML: {len(texto_html)} caracteres")
```

### **Desafio do MÃ³dulo - Criando um Sistema Completo**

Vamos criar um sistema que escolhe automaticamente o melhor splitter para cada tipo de texto:

```python
# ğŸ¯ DESAFIO: SISTEMA INTELIGENTE DE SPLITTING

def escolher_splitter_inteligente(texto):
    """Escolhe o melhor splitter baseado no tipo de texto"""
    
    # Detectando tipo de texto
    if texto.strip().startswith('<html') or '<body>' in texto:
        print("ğŸŒ Detectado HTML - usando HTML Splitter")
        return HTMLTextSplitter(chunk_size=200, chunk_overlap=30)
    
    elif '#' in texto[:100] and '\n' in texto:
        print("ğŸ“ Detectado Markdown - usando Markdown Splitter")
        return MarkdownTextSplitter(chunk_size=200, chunk_overlap=30)
    
    else:
        print("ğŸ“„ Texto normal - usando Recursive Splitter")
        return RecursiveCharacterTextSplitter(
            chunk_size=200,
            chunk_overlap=30,
            separators=["\n\n", "\n", ". ", "! ", "? ", " ", ""]
        )

# Testando o sistema inteligente
textos_teste = [
    ("Texto normal", "Este Ã© um texto normal sem formataÃ§Ã£o especial."),
    ("Markdown", "# TÃ­tulo\n\nEste Ã© um texto em markdown."),
    ("HTML", "<html><body><h1>TÃ­tulo</h1><p>Texto HTML</p></body></html>")
]

print("ğŸ¯ TESTANDO SISTEMA INTELIGENTE DE SPLITTING")
print("=" * 50)

for nome, texto in textos_teste:
    print(f"\nğŸ” Testando: {nome}")
    splitter = escolher_splitter_inteligente(texto)
    chunks = splitter.split_text(texto)
    print(f"   ğŸ“„ PedaÃ§os criados: {len(chunks)}")
    print(f"   ğŸ“ Tamanho mÃ©dio: {sum(len(chunk) for chunk in chunks) / len(chunks):.0f} chars")

print("\nğŸ‰ Sistema inteligente funcionando perfeitamente!")
```
